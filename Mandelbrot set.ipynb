```python
import cmath
import math
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
```

```python
def mandelFunction(z,c):
    return z**2 + c

def radius(c):
    return math.sqrt(c.real**2 + c.imag**2)
```

```python
def check(c):
    z = 0
    #where c is a complex number
    i=1
    while radius(z)<=2 and i<=100:
        z = mandelFunction(z,c)
        i+=1
    return i
```

```python
def plot(x1,y1,x2,y2,n,colorMap):
    a = (x2 - x1)/n
    b = (y2 - y1)/n
    xs = []
    ys = []
    colours = []
    for i in range(0,n+1):
        for j in range(0,n+1):
            x = x1 + a*i
            y = y1 + b*j
            c = complex(x,y)
            xs.append(x)
            ys.append(y)
            colours.append(check(c))
    figure(num=None, figsize=(5, 5), dpi=80, facecolor='w', edgecolor='k')
    plt.scatter(xs, ys, c=colours, cmap = colorMap, marker='o', s = 10)
#     plt.figure(figsize=(100,100))
```
