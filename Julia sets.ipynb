```python
import cmath
import math
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
import random
```

```python
def juliaIIM(c,limit):
    xs = []
    ys = []
    z = (1+cmath.sqrt(1-4*c))/2
    i=1
    while i < limit:
        i+=1
        x = z.real
        y = z.imag
        xs.append(x)
        ys.append(y)
        if random.choice([0,1]) == 0:
            z = cmath.sqrt(z - c)
        else:
            z = - cmath.sqrt(z - c)
    plt.plot(xs,ys,'o', markersize = 1)
```

```python
juliaIIM(0-0,100000)
```

```python
def juliaPlot(x1,y1,x2,y2,n,colorMap, c):
    a = (x2 - x1)/n
    b = (y2 - y1)/n
    xs = []
    ys = []
    colours = []
    for i in range(0,n+1):
        for j in range(0,n+1):
            x = x1 + a*i
            y = y1 + b*j
            z0 = complex(x,y)
            xs.append(x)
            ys.append(y)
            colours.append(julia(z0,c))
    figure(num=None, figsize=(5, 5), dpi=80, facecolor='w', edgecolor='k')
    plt.scatter(xs, ys, c=colours, cmap = colorMap, marker='o', s = 10)
```

```python
juliaPlot(-2,-2,2,2,100,'spring_r',-0.4 + 0.6j)
```
