```python
import cmath
import math
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
import random
```

# Using IIM method for boundary of julia set

```python
def juliaIIM(c,limit):
    xs = []
    ys = []
    z = (1+cmath.sqrt(1-4*c))/2
    i=1
    while i < limit:
        i+=1
        x = z.real
        y = z.imag
        xs.append(x)
        ys.append(y)
        if random.choice([0,1]) == 0:
            z = cmath.sqrt(z - c)
        else:
            z = - cmath.sqrt(z - c)
    figure(num=None, figsize=(5,5), dpi=80, facecolor='w', edgecolor='k')
    plt.plot(xs,ys,'o', markersize = 1)
```

```python
juliaIIM(-0.1+0.651j,10000)
```

# Implementing wiki pseudocode

```python
def check(x,y,c,limit):
    i = 0
    limit = 1000
    while x**2 + y**2 < 4 and i<limit:
        i+=1
        (x, y) = (x**2 - y**2 + c.real, 2*x*y  +  c.imag)
    return i

def juliaPseud(c, limit, x1,x2,y1,y2,n):
    a = (x2-x1)/n
    b = (y2-y1)/n
    xs = []
    ys = []
    colours = []
    for i in range(0,n+1):
        for j in range(0,n+1):
            x = x1 + a*i
            y = y1 + b*j
            xs.append(x)
            ys.append(y)
            colours.append(check(x,y,c,limit))
    figure(num=None, figsize=(5, 5), dpi=80, facecolor='w', edgecolor='k')
    plt.scatter(xs, ys, c=colours, cmap = 'twilight_shifted_r', marker='o', s = 1)
```

```python
juliaPseud(-1.4,1000,-1, 1, -1, 1,400)
```
